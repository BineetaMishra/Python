```python
import matplotlib.pyplot as plt
import numpy as np # for matplotlib library
import seaborn as sns
import pandas as pd # for seaborn library
```

## Using Matplotlib

```python
arr1= np.arange(5,15)
arr1
```

```python
arr2 = arr1** 2
arr2
```

```python
#do plot#creating the fig
fig = plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(arr1,arr2)
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr2')
axes.set_title('Arr1 vs Arr2')
```

```python
#do plot#creating the fig
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.4])
axes1.plot(arr1,arr2)
axes1.set_xlabel('Arr1')
axes1.set_ylabel('Arr2')
axes1.set_title('Arr1 vs Arr2')

axes2.plot(arr2,arr1)
axes2.set_xlabel('Arr2')
axes2.set_ylabel('Arr1')
axes2.set_title('Arr2 vs Arr1')
```

```python
#import pyplot sub part
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr1= np.arange(5,15)
arr1
```

```python
arr2 = arr1** 2
arr2
```

```python
#creating the fig
fig = plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8]) #length, bottom,width , height (range : 0 to 1)
```

```python
#do plot#creating the fig
fig = plt.figure()
axes= fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(arr1,arr2)
axes.set_xlabel('Arr1')
axes.set_ylabel('Arr2')
axes.set_title('Arr1 vs Arr2')
```

```python
#do plot#creating the fig
fig = plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.4])
axes1.plot(arr1,arr2)
axes1.set_xlabel('Arr1')
axes1.set_ylabel('Arr2')
axes1.set_title('Arr1 vs Arr2')

axes2.plot(arr2,arr1)
axes2.set_xlabel('Arr2')
axes2.set_ylabel('Arr1')
axes2.set_title('Arr2 vs Arr1')
```

```python
#do plot#creating the fig
fig = plt.subplots(nrows=2, ncols=2)
#axes1.plot(arr1,arr2)
#axes1.set_xlabel('Arr1')
#axes1.set_ylabel('Arr2')
#axes1.set_title('Arr1 vs Arr2')

#axes2.plot(arr2,arr1)
#axes2.set_xlabel('Arr2')
#axes2.set_ylabel('Arr1')
#axes2.set_title('Arr2 vs Arr1')

plt.tight_layout() #to remove over lapping
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,8), dpi=100)
axes[0][0].plot(arr1,arr2)
axes[0][0].set_xlabel('Arr1')
axes[0][0].set_ylabel('Arr2')
axes[0][0].set_title('Arr1 vs Arr2')
axes[0][1].plot(arr2,arr1)
axes[0][1].set_xlabel('Arr2')
axes[0][1].set_ylabel('Arr1')
axes[0][1].set_title('Arr2 vs Arr1')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
pokemon_data= pd.read_csv('')
```

```python
pokemon_data.head()
```

LMPLOT gives scatter and regression plot

```python
sns.lmplot(x='Spped Point',
          y='Defense Point',
          data= pokemon_data)
plt.show()
#Regression Line and Confidense Interval
```

```python
sns.lmplot(x='Spped Point',
          y='Defense Point',
           col= 'Type1',
          data= pokemon_data)
plt.show()
```

```python
pokemon_data['Type1'].value_counts()
```

```python
#hue adds color to the graph
sns.lmplot(x='Spped Point',
          y='Defense Point',
           hue= 'Type1',
          data= pokemon_data)
plt.show()
```

```python
sns.lmplot(x='Spped Point',
          y='Defense Point',
           hue= 'Type2',
           col='Type1',
          data= pokemon_data)
plt.show()
```

diff btw col and hue
hue will add another dimension

# JOINT PLOT

```python
sns.jointlot(x='Spped Point',
          y='Defense Point',
           kind='reg',
          data= pokemon_data)
plt.show()
```

```python
sns.pairlot(x='Spped Point',
            size=5, aspect=0.9
          x_vars=['Defense Point'],
        y_vars= ['Total Point'],
           kind='reg',
          data= pokemon_data)
plt.show()
```

# Categorical Data Analysis

we need one axis to be cateregorical data

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#
sns.set(style='whitegrid',
       color_codes= True)
```

```python
sns.stripplot(x='day',
             y='tip',
             data= tips)
plt.show()
```

```python
plt.supplots(figsize(9,10))
```

```python
#SWARMPLOT
plt.subplots(figsize=(9,10))
sns.swarmplot(x='day',
             y='tip',
             data= tips)
plt.show()
```

```python
#SWARMPLOT
plt.subplots(figsize=(9,10))
sns.swarmplot(x='day',
             y='tip',
             data= tips,
             hue='time')
plt.show()
```

```python
#BOXPLOT
plt.subplots(figsize=(9,10))
sns.boxplot(x='day',
           y='tip',
           data=tips)
#Outliers, whiskers, interquartile ranges
```

```python
#VIOLIN PLOTs
#BOXPLOT
plt.subplots(figsize=(9,10))
sns.violinplot(x='day',
           y='tip',
           data=tips)
```

```python
#BOXPLOT
plt.subplots(figsize=(9,10))
sns.violinplot(x='day',
           y='tip',
           data=tips,
              inner=None)
sns.swarmplot(x='day',
           y='tip',
           data=tips,
             color='w')
plt.show()
```

```python
#Factor Plot
sns.fac(x='day',
           y='tip',
           data=tips,
       col
       size
       hue
       kind='box')
```

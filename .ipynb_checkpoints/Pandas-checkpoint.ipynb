```python
import numpy as np
import pandas as pd
```

```python
students = pd.DataFrame({'id_number' : [204, 202, 264, 136, 238],
                               'Name' : ['James Smith',
                                        'Maria Rodriguez',
                                        'David Smith', 
                                        'Patricia Garcia',
                                        'Jessica Hernandez'],
                               'Age' : [26, 22 ,24, 21, 25],
                            'Gender' : ['M', 'F', 'M', 'F', 'F'],
                    
                        })
```

```python
students
```

```python
professors = pd.DataFrame({'Prof_name' : ['Thomas Smith',
                                          'Margaret Garcia',
                                          'James Johnson', 
                                          'Richard Miller',
                                          'Margaret MOORE'],
                      'id_number' : [204, 264, 108, 238, 136],
                     'Department' : ['ECE', 'EE' ,'CSE', 'ME', 'CE'],
                      
                         })
```

```python
professors
```

```python
pd.concat([students, professors], sort=True)
```

```python
pd.concat([students, professors], sort=True, ignore_index=True)
```

```python
#whichever column is common tat will be merged
pd.merge(students, professors)
```

```python
#all the id of student df is taken and professor is NAN
pd.merge(students, professors, how='left')
```

```python
pd.merge(students, professors, how='right')
```

```python
#default merging
pd.merge(students, professors, how='inner')
```

Concat is packing the data appending the data together.
Ignore_index to get the normal indexing
Merge is common col based on that the 2 df will be merged. which ID is not merging it will be removed. But in case of how=true, the id will be taken.
